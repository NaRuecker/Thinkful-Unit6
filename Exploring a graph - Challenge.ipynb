```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline

import networkx as nx
```

```python
G = nx.read_gml('graph.gml')
```

```python
#Drawing the graph
plt.figure(figsize=(10,10))
nx.draw_networkx(G,
                 pos=nx.spring_layout(G), # Positions the nodes relative to each other
                 node_size=100, 
                 cmap=plt.cm.Blues, # The color palette to use to color the nodes.
                 node_color=range(len(G)) #The number of shades of color to use.
                 )
plt.axis('off')
plt.show()
```

1. How many nodes ? 77
2. How many edges ? 254

```python
print("This graph has {} nodes and {} edges.".format(G.number_of_nodes(),G.number_of_edges()))
```

 3. Is this graph directional? Yes.

```python
# the initial graph after reading the gml file is not directed,
print("Is the graph directed: ",G.is_directed())
DG=G.to_directed()
print("Is the graph directed after converting: ",DG.is_directed())
#Drawing the graph
plt.figure(figsize=(15,15))
nx.draw_networkx(DG,
                 pos=nx.spring_layout(G), # Positions the nodes relative to each other
                 node_size=100, 
                 cmap=plt.cm.Blues, # The color palette to use to color the nodes.
                 node_color=range(len(G)) #The number of shades of color to use.
                 )
plt.axis('off')
plt.show()
```

4. Which node has the highest degree?

```python
degree_seq=sorted(G.degree(), key=lambda x: x[1], reverse=True)
degree_seq[0]
```

5. Average clustering coefficient (if possible)

```python
cc=[]
for k,v in nx.clustering(G).items():
    cc.append(v)
np.mean(cc)
```

6. The nodes with the three highest betweenness centrality scores, and their scores:

```python
betweenness_seq=sorted(nx.betweenness_centrality(G).items(), key=lambda x: x[1], reverse=True)
betweenness_seq[0:3]
```

7. Also experiment with various ways to draw the network.

```python
plt.figure(figsize=(8,8))
nx.draw_networkx(DG,
                 pos=nx.circular_layout(G), # Positions the nodes relative to each other
                 node_size=100, 
                 cmap=plt.cm.Blues, # The color palette to use to color the nodes.
                 node_color=range(len(G)) #The number of shades of color to use.
                 )
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(8,8))
nx.draw_networkx(DG,
                 pos=nx.random_layout(G), # Random
                 node_size=100, 
                 cmap=plt.cm.Blues, # The color palette to use to color the nodes.
                 node_color=range(len(G)) #The number of shades of color to use.
                 )
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(8,8))
nx.draw_networkx(DG,
                 pos=nx.shell_layout(G), # Random
                 node_size=100, 
                 cmap=plt.cm.Blues, # The color palette to use to color the nodes.
                 node_color=range(len(G)) #The number of shades of color to use.
                 )
plt.axis('off')
plt.show()
```
